```python
class Library:
    
    def __init__(self, listOfBooks):
        self.books = listOfBooks
    
    def displayAvailableBooks(self):
        if (len(self.books))> 0:
            print("\nBooks present in this library are: ")
            for book in self.books:
                print("*",book)
                
        else:
            print("Sorry! There are no books available at this time. Kindly wait for the books to be available")
    
    def borrowBook(self, nameOfBook):
        if nameOfBook in self.books:
            print(f"You have been issued {nameOfBook}.Please keep it safe and return it within 30 days otherwise fine will be charged")
            self.books.remove(nameOfBook)
        else:
            print("Sorry! This book is either not present or it has been taken by someone else. Please wait until the book is available")
            
    def returnBook(self, nameOfBook):
        self.books.append(nameOfBook)
        print("Thanks for returning this book! Hope you enjoyed reading it. Have a great day ahead!")

class Student:
    def requestBook(self):
            self.book = input("Enter the name of the book you want to borrow: ")
            self.book1 = self.book.capitalize()
            return self.book1
        
    def returnBook(self):
        self.book = input("Enter the name of the book you want to return or add: ")
        self.book1 = self.book.capitalize()
        return self.book1
    
centralLibrary = Library(["English", "Hindi", "Maths"])
student = Student()

print("*****Welcome to the Student Library*****")

explore1 = "Yes"
explore = explore1.upper()

while explore == "YES":
    msg = '''\n\tPlease choose an option:
            1. List all the books
            2. Request a book
            3. Add/Return a book
            4. Exit the Library
        '''
    print(msg)
    a = int(input("Enter your choice: "))
    if a == 1:
        centralLibrary.displayAvailableBooks()
        
    elif a == 2:
        centralLibrary.borrowBook(student.requestBook())
        explore1 = input("Do you want to explore more ? Yes or No: ")
        explore = explore1.upper()

        if explore == "NO":
            print("\nThanks for choosing Student Library. Have a great day ahead!")
            exit()
            
    elif a == 3:
        centralLibrary.returnBook(student.returnBook())
        explore1 = input("Do you want to explore more ? Yes or No: ")
        explore = explore1.upper()
        
        if explore == "NO":
            print("\nThanks for choosing Student Library. Have a great day ahead!")
            exit()
            
    elif a == 4:
        print("\nThanks for choosing Student Library. Have a great day ahead!")
        explore = "NO"
        
    else:
        print("Invalid Choice!")
```
